```python
# imports
import webbrowser
import time
from pynput.keyboard import Key, Controller
import pandas as pd
import numpy as np
```

```python
# Variables/Constants
url = 'http://localhost/cont-auth/extern_input/same_origin.html'
iterations = 10
time_to_sleep_sec = 0.1

file = "./DSL-StrongPasswordData.csv"

keyboard = Controller()
```

```python
# get dataframe
df = pd.read_csv(file)

# get row
row = df.iloc[0]
row

# store Hold values
h_period = row['H.period']
h_t = row['H.t']
h_i = row['H.i']
h_e = row['H.e']
h_five = row['H.five']
h_shift_r = row['H.Shift.r']
h_o = row['H.o']
h_a = row['H.a']
h_n = row['H.n']
h_l = row['H.l']
h_return = row['H.Return']


# store UpDown values
ud_period_t = row['UD.period.t']
ud_t_i = row['UD.t.i']
ud_i_e = row['UD.i.e']
ud_e_five = row['UD.e.five']
ud_five_R = row['UD.five.Shift.r']
ud_R_o = row['UD.Shift.r.o']
ud_o_a = row['UD.o.a']
ud_a_n = row['UD.a.n']
ud_n_l = row['UD.n.l']
ud_l_return = row['UD.l.Return']
```

```python
webbrowser.open(url)  # open webpage in browser
time.sleep(3)  # wait for user to be ready (select input field)

# .tie5Roanl

keyboard.press('.')
time.sleep(h_period)  # H.period
keyboard.release('.')

time.sleep(ud_period_t)  # UD.period.t

keyboard.press('t')
time.sleep(h_t)  # H.t
keyboard.release('t')

time.sleep(ud_t_i)  # UD.t.i

keyboard.press('i')
time.sleep(h_i)  # H.i
keyboard.release('i')

time.sleep(ud_i_e)  # UD.i.e

keyboard.press('e')
time.sleep(h_e)  # H.e
keyboard.release('e')

time.sleep(ud_e_five)  # UD.e.five

keyboard.press('5')
time.sleep(h_five)  # H.five
keyboard.release('5')

time.sleep(ud_five_R)  # UD.five.Shift.r

keyboard.press('R')
time.sleep(h_shift_r)  # H.Shift.r
keyboard.release('R')

time.sleep(ud_R_o)  # UD.Shift.r.o

keyboard.press('o')
time.sleep(h_o)  # H.o
keyboard.release('o')

time.sleep(ud_o_a)  # UD.o.a

keyboard.press('a')
time.sleep(h_a)  # H.a
keyboard.release('a')

time.sleep(ud_a_n)  # UD.a.n

keyboard.press('n')
time.sleep(h_n)  # H.n
keyboard.release('n')

time.sleep(ud_n_l)  # UD.n.l

keyboard.press('l')
time.sleep(h_l)  # H.l
keyboard.release('l')

time.sleep(ud_l_return)  # UD.l.Return

keyboard.press(Key.enter)
time.sleep(h_return)  # H.Return
keyboard.release(Key.enter)
```
